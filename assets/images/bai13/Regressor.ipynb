```python
from sklearn.tree import DecisionTreeRegressor
import numpy as np
```

```python
X = np.array([2,5,6,7, 10,12,  15,18,20,21,  23,24,25,26,27, 32, 34,35,38]).reshape(-1,1)
y = np.array([0,0,0,0, 5,20,  100,100,100,100,  58,57,55, 54,53, 15,  0,0,0])
```

```python
import matplotlib.pyplot as plt
plt.scatter(X,y)
plt.xlabel('Drug(mg)')
plt.ylabel('Effective(%)')
```

```python
regressor = DecisionTreeRegressor(random_state=0)
```

```python
regressor.fit(X,y)
```

```python
import graphviz
from sklearn import tree

dot_data = tree.export_graphviz(regressor, out_file=None, 
                                rounded=True,
                                filled=True)

graph = graphviz.Source(dot_data, format="png")
graph
```

```python
y_pred = regressor.predict(X)
```

```python
y_pred
```

```python
plt.scatter(X,y)
plt.xlabel('Drug(mg)')
plt.ylabel('Effective(%)')
plt.plot(X,y_pred,'r')
```
